```python
df=pd.read_csv("/content/drive/MyDrive/DAY 1/student_performance.csv")
```

```python
print("============================================================")
print(" STUDENT PERFORMANCE EXPLORER REPORT")
print("============================================================")


print("[1] DATASET OVERVIEW")
print("------------------------------------------------------------")
print("Total Students:",df.shape[0])
print("Total columns:",df.shape[1])
print("Total Departments:",len(df['department'].unique()))

```

```python
print("[2] DEPARTMENT-WISE STUDENT COUNT")
print("------------------------------------------------------------")

print(df.groupby('department')['name'].count())
```

```python
print("[2] DEPARTMENT-WISE STUDENT COUNT")
print("------------------------------------------------------------")

print(df['department'].value_counts())
```

```python
print("[3] SCORE ANALYSIS")
print("------------------------------------------------------------")
print("Highest Math Score:",df['math_score'].max())
print("Lowest Math Score:",df['math_score'].min())
print(f"Average Math Score: {df['math_score'].mean():.2f}")
print(f"Average programming score: {df['programming_score'].mean():.2f}")
print(f"Average Attendance: {df['attendance_percentage'].mean():.2f}%")
```

```python
print("[4] GENDER-WISE AVERAGE SCORES")
print("-"*40)
df['total_score']=df['math_score']+df['english_score']+df['programming_score']+df['science_score']
gender_scores=df.groupby('gender')['total_score'].mean()
print(gender_scores)
```

```python
print("[5] TOP 5 PERFORMERS")
print("-"*40)
top5=df.sort_values('total_score',ascending=False).head(5)
for index,row in top5.iterrows():
  print(f"{row['name']:<20} | {row['department']:<18} | Score: {row['total_score']}")
```

```python
print("[6] STUDENT WHO NEED ATTENTION")
print("-"*40)
low_attendance=df[df['attendance_percentage']<75][['name','department','attendance_percentage']]
if len(low_attendance)>0:
  print(low_attendance.to_string(index=False))
else:
  print("All Students have attendance above 75%")

print("\n" + "="*60)
print("Report Complete")
print("="*60)
```
