```python
cs_students = df[df['department'] =='Computer Science']

print("Number of Computer Science students:{len(cs_students)}")

print("\nComputer Science Students:")
print(cs_students[['name','math_score','programming_score','attendance_percentage']])
```

```python
print("Computer Science Students above 90% attendance students")

cs_students_attendance = cs_students[cs_students['attendance_percentage']>90]

print(cs_students_attendance[['name','attendance_percentage']])
```

```python
df['total_score']=df['math_score']+df['english_score']+df['programming_score']+df['science_score'];
print("Students above 90%:")
toppers=df[df['total_score']/4 > 90]
print(toppers[['name','department']])

```

```python
dept_avg=df.groupby('department')['total_score'].mean().round(3)

print("Department wise average score")
print(dept_avg)
```

```python
print("Descending order of department :")
desc = dept_avg.sort_values(ascending=False)
print(desc)
```
